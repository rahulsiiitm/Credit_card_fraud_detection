```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os

# Add 'src' to path
sys.path.append(os.path.abspath(os.path.join('..')))

df = pd.read_csv('../data/raw/creditcard.csv')
```

```python
# Count the classes
counts = df['Class'].value_counts()
print(counts)
print(f"\nFraud Percentage: {(counts[1]/len(df))*100:.3f}%")

# Plot
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df, palette='viridis', edgecolor=sns.color_palette('dark', n_colors=1)[0])
plt.title('Class Distribution \n (0: No Fraud || 1: Fraud)', fontsize=14)
plt.show()
```

```python
# Let's see which features correlate most with Fraud (Class)
plt.figure(figsize=(20,10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm_r', annot_kws={'size':20})
plt.title("Imbalanced Correlation Matrix", fontsize=14)
plt.show()
```
